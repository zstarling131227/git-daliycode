```python
fig,axes=plt.subplots(2,2,sharex=True,sharey=True)
```

```python
fig,axes=plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),bins=10,color="r",alpha=0.7)   #  bins是分组个数
plt.subplots_adjust(wspace=0,hspace=0)
```

```python
data=np.random.randn(30).cumsum()    #  累加和
data
plt.plot(data,"ko--",label="Default")
```

```python
plt.plot(data,"k--",label="Default")
plt.plot(data,"k-",drawstyle="steps-post",label="steps-post")
plt.legend(loc="best")  #legend方法有多个其他的位置参数loc。loc参数告诉matplotlib在哪里放置图表。如果你不挑剔，’best’是一个好选项，它会自动选择最合适的位置。
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
tick=ax.set()
ax.plot(np.random.randn(1000).cumsum(),label="one")
tick=ax.set_xticks([0,250,500,750,1000]) # set_xticks表示在数据范围内设定刻度的位置
# label=ax.set_xticklabels(["one","two","three","four","five"],rotation=30,fontsize="xx-large") # 使用set_xticklabels为标签赋值
plt.legend(loc="best")
plt.title(" titicks")
```
